```python
# Handle ranges in world_rank column and convert them to integers

def convert_world_rank(rank):
    if '-' in rank:
        start, end = rank.split('-')
        return (int(start) + int(end)) // 2
    elif '=' in rank:
        return int(rank.replace('=', ''))
    else:
        return int(rank)

df4['world_rank'] = df4['world_rank'].apply(convert_world_rank)
df4['teaching'] = df4['teaching'].astype(float)
df4['research'] = df4['research'].astype(float)
df4['citations'] = df4['citations'].astype(float)
```
